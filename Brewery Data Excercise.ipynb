```python
idls = []
namels = []
createls = []
statusls = []
organicls = []
retiredls = []
descls = [] 
```

```python
for beer in beers:
        
    #get the value of each key and if the key doesn't exist, set a variable to be None
    try: kid= beer['id']
    except: kid = None
               
    try: dname = beer['nameDisplay']
    except: dname = None
        
    try: create = beer['createDate']
    except: create = None
    
    try: stat = beer['status']
    except: stat = None
    
    try: org = beer['isOrganic']
    except: org = None
            
    try: ret = beer['isRetired']
    except: ret = None
     
    try: ldesc = beer['style']['description']
    except: ldesc = None
            
    idls.append(kid)
    namels.append(dname)
    createls.append(create)
    statusls.append(stat)
    organicls.append(org)
    retiredls.append(ret)
    descls.append(ldesc)
       
```

```python
print(idls[:25])
print(namels[:25])
print(createls[:25])
print(statusls[:25])
print(organicls[:25])
print(retiredls[:25])
print(descls[:25])
```

```python
beerList = list(zip(idls,namels,createls,statusls,organicls,retiredls,descls))

#names for columns in data frame
colnames = ['id','Name','CreateDate','Status', 'Organic','Retired','Description']
```

```python
df = pd.DataFrame(beerList, columns=colnames)

df
```
