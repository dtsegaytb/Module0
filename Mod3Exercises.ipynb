### Exercise 7:

Using the dataframe from the previous exercise, pick a method (Standard Deviation or Interquartile Range) and remove the outliers from the "FuelEconomy" column.

```python
#Standard Deviation
meanFuelEco = cdrop['FuelEconomy'].mean()
stdFuelEco = cdrop['FuelEconomy'].std()
topFuelEco = meanFuelEco + stdFuelEco * 1.96
botFuelEco = meanFuelEco - stdFuelEco * 1.96

newdf = cdrop.copy()#Copy original
newdf = newdf.drop(newdf[newdf['FuelEconomy'] > topFuelEco].index)
newdf = newdf.drop(newdf[newdf['FuelEconomy'] < botFuelEco].index)

newdf.head()
```
